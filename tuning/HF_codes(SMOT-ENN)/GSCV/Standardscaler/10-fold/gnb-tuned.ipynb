```python
import warnings
warnings.filterwarnings("ignore")
import time
to=time.clock()
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler 
scale=StandardScaler()
heart="../input/heart-failure/heart_failure.csv"
data=pd.read_csv (heart)
x=data.drop('DEATH_EVENT',axis=1)
x=scale.fit_transform(x)
y=data['DEATH_EVENT']
from imblearn.combine import SMOTEENN
ada = SMOTEENN()
x,y = ada.fit_resample(x, y)

grid={
    'var_smoothing':[0.000000001,0.000001,0.1,1]
}
model=GaussianNB()
rdm = GridSearchCV(model,grid,cv=10)
rdm.fit(x,y)
print("GNB accuracy:",rdm.best_score_)
print("GNB best params:",rdm.best_params_)
t1=time.clock()
print(t1-to)
```
