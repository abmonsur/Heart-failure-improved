```python
import warnings
warnings.filterwarnings("ignore")
import time
to=time.clock()
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler 
scale=StandardScaler()
heart="../input/heart-failure/heart_failure.csv"
data=pd.read_csv (heart)
x=data.drop('DEATH_EVENT',axis=1)
x=scale.fit_transform(x)
y=data['DEATH_EVENT']
from imblearn.combine import SMOTEENN
ada = SMOTEENN()
x,y = ada.fit_resample(x, y)
grid={
    'n_estimators':[50,100,200],
    'min_samples_split':[1,2,5,20],
    'min_samples_leaf':[1,2,5,20],
    'bootstrap':[True,False],
}
model=RandomForestClassifier()
rdm = RandomizedSearchCV(model,grid,cv=10)
rdm.fit(x,y)
print("RFC accuracy:",rdm.best_score_)
print("RFC best params:",rdm.best_params_)
t1=time.clock()
print(t1-to)
```
